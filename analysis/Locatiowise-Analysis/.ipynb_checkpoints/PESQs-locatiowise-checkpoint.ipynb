```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams


# === CONFIGURE FOR ACADEMIC QUALITY PLOTS ===
rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 13,
    "axes.titlesize": 14,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "pdf.fonttype": 42,  # Editable fonts
})
# Read the CSV file
data_file = "data/super-tor.csv"
df = pd.read_csv(data_file)

# Filter out rows where caller_pesq_mos or callee_pesq_mos are NaN
df = df.dropna(subset=['caller_pesq_mos', 'callee_pesq_mos'])

# Define the updated label map for location names
label_map = {
    "1Frank-Lond":          "Frankfurt - London",
    "2sydney-frank":        "Sydney - Frankfurt",
    "3Bangl-nyc":           "Bangalore - New York",
    "4singapore-nyc":       "Singapore - New York",
    "5sans-amster":         "San Francisco - Amsterdam",
    "6nyc-syd":             "New York - Sydney",
    "7CapeTown-Seoul":      "Cape Town - Seoul",
    "8Jakarta-Mexico":      "Jakarta - Mexico",
    "9Osaka-UAE":           "Osaka - UAE",
    "10SauPaul-malaysia":   "SÃ£o Paulo - Malaysia",
    "11TelAviv-Sydney3":    "Tel Aviv - Sydney",
    "Oregon-California":    "Oregon-California",
    "Paris-London":         "Paris-London",
    "Stockholm-Frankfurt":  "Stockholm-Frankfurt", 
    "Stockholm-Frankfurt":  "Virginia-Ohio"
}

# Only keep rows for the locations in label_map
df = df[df['location'].isin(label_map.keys())]

# Group the data by location and codec to count the number of calls per codec in each location
grouped = df.groupby(['location', 'codec']).size().reset_index(name='num_calls')

# Group the data by location for the CDF plot
df_grouped = df.groupby('location')

# Create the plot for CDF
plt.figure(figsize=(10, 6))

# Iterate over each group (location)
for location, group in df_grouped:
    # Sort the PESQ scores
    caller_pesq_sorted = np.sort(group['caller_pesq_mos'])
    callee_pesq_sorted = np.sort(group['callee_pesq_mos'])
    
    # Merge the CDF values for caller and callee PESQ scores
    merged_pesq_sorted = np.sort(np.concatenate((caller_pesq_sorted, callee_pesq_sorted)))
    
    # Calculate the CDF for merged values
    merged_pesq_cdf = np.arange(1, len(merged_pesq_sorted) + 1) / len(merged_pesq_sorted)

    # Use the label_map to get the descriptive location names
    location_name = label_map.get(location, location)  # Default to the original location code if not found in the map

    # Plot the CDF with a thicker line and distinct style
    plt.plot(merged_pesq_sorted, merged_pesq_cdf, label=f'{location_name}', linewidth=2)

    # Calculate the percentage of calls greater than 3.0 for merged PESQ values
    greater_than_3_count = np.sum(merged_pesq_sorted > 3.0)
    total_calls = len(merged_pesq_sorted)
    percentage_greater_than_3 = (greater_than_3_count / total_calls) * 100

    # Calculate average, min, and max PESQ values
    avg_pesq = np.mean(merged_pesq_sorted)
    min_pesq = np.min(merged_pesq_sorted)
    max_pesq = np.max(merged_pesq_sorted)

    # Print the statistics for each location
    print(f"Location: {location_name}")
    print(f"  Number of instances: {total_calls}")
    print(f"  Percentage of calls with PESQ > 3.0: {percentage_greater_than_3:.2f}%")
    print(f"  Average PESQ: {avg_pesq:.2f}")
    print(f"  Min PESQ: {min_pesq:.2f}")
    print(f"  Max PESQ: {max_pesq:.2f}")
    
    # Print the number of calls per codec
    location_codec_data = grouped[grouped['location'] == location]
    for _, row in location_codec_data.iterrows():
        codec_name = row['codec']
        num_calls = row['num_calls']
        print(f"    Codec: {codec_name}, Number of calls: {num_calls}")

    print("-" * 50)

# Add labels and title
plt.xlabel('PESQ MOS Score', fontsize=12) #, fontweight='bold')
plt.ylabel('CDF', fontsize=12) #, fontweight='bold')
# plt.title('CDF of Merged Caller and Callee PESQ MOS Scores Location-Wise', fontsize=14) #, fontweight='bold')

# Improve legend appearance and placement
plt.legend(loc='lower right', fontsize=9, frameon=False, title="Location", title_fontsize=12)

# Add grid lines with lower opacity for a cleaner look
plt.grid(True, linestyle='--', linewidth=0.8, alpha=0.6)

# Save the plot as a PDF file (suitable for conference submission)
plt.tight_layout()
plt.savefig("merged_pesq_cdf_location_wise.pdf", dpi=300, bbox_inches="tight", transparent=True)

# Show the plot
plt.show()
```
