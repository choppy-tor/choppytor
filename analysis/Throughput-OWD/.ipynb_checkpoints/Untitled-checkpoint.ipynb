```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV file
file_path = "../NewData/Non-Tor/combined_results.csv"  # Change this to your actual file path
df = pd.read_csv(file_path, sep=",")  # Use tab separator if that's what your file uses

# Convert PESQ and jitter columns to numeric, coercing empty strings to NaN
columns_to_convert = [
    "caller_pesq_mos", "callee_pesq_mos",
    "caller_jitter_mean", "callee_jitter_mean",
    "caller_jitter_std_dev", "callee_jitter_std_dev"
]
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Drop rows where all relevant values are NaN (for each plot separately)
pesq_data = df["caller_pesq_mos"].dropna().tolist() + df["callee_pesq_mos"].dropna().tolist()
jitter_mean_data = df["caller_jitter_mean"].dropna().tolist() + df["callee_jitter_mean"].dropna().tolist()
jitter_std_data = df["caller_jitter_std_dev"].dropna().tolist() + df["callee_jitter_std_dev"].dropna().tolist()

def plot_cdf(data, title, xlabel, filename):
    data_sorted = np.sort(data)
    cdf = np.arange(1, len(data_sorted) + 1) / len(data_sorted)
    plt.figure()
    plt.plot(data_sorted, cdf, marker='.', linestyle='none')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("CDF")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(filename)
    print(f"Saved: {filename}")

# Plot 1: PESQ MOS CDF
plot_cdf(pesq_data, "CDF of PESQ MOS (Caller + Callee)", "PESQ MOS", "pesq_mos_cdf.png")

# Plot 2: Jitter Mean CDF
plot_cdf(jitter_mean_data, "CDF of Jitter Mean (Caller + Callee)", "Jitter Mean (ms)", "jitter_mean_cdf.png")

# Plot 3: Jitter Std Dev CDF
plot_cdf(jitter_std_data, "CDF of Jitter Std Dev (Caller + Callee)", "Jitter Std Dev (ms)", "jitter_std_dev_cdf.png")
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio

# === CONFIGURE ===
csv_path = "data/super-tor.csv"  # Adjust path if needed
output_html = "pesq_cdf_by_location.html"
output_png = "pesq_cdf_by_location.png"
separator = ","

# === LOAD DATA ===
df = pd.read_csv(csv_path, sep=separator)

# === INITIALIZE PLOTLY FIGURE ===
fig = go.Figure()

# === GROUP AND FILTER LOCATIONS ===
locations = df["location"].dropna().unique()

for location in locations:
    group = df[df["location"] == location]
    pesq_mos = pd.concat([group["caller_pesq_mos"], group["callee_pesq_mos"]], ignore_index=True).dropna()

    if len(pesq_mos) <= 100:
        continue  # Skip locations with <= 100 entries

    pesq_sorted = np.sort(pesq_mos)
    cdf = np.arange(1, len(pesq_sorted) + 1) / len(pesq_sorted)

    fig.add_trace(go.Scatter(
        x=pesq_sorted,
        y=cdf,
        mode='lines+markers',
        name=location
    ))

# === FINALIZE PLOT ===
fig.update_layout(
    title="CDF of PESQ MOS by Location (Only Locations with >100 Samples)",
    xaxis_title="PESQ MOS",
    yaxis_title="CDF",
    legend_title="Location",
    template="plotly_white"
)

# === SAVE HTML FILE ===
pio.write_html(fig, file=output_html, auto_open=False)

# # === SAVE STATIC IMAGE (requires kaleido) ===
# try:
#     pio.write_image(fig, file=output_png, format='png', width=1200, height=800, scale=2)
# except Exception as e:
#     print(f"Could not save PNG. Error: {e}")
```
