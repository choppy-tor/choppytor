```python
import os
import pandas as pd
import numpy as np
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm

# === CONFIGURATION ===
root_dir = "/home/jithin/data/IMC/NewData/owd/tor/"  # UPDATE THIS
output_file = "delay_statistics_summary-Tor.csv"

def process_csv(file_info):
    iteration, full_path = file_info
    try:
        df = pd.read_csv(full_path)
        delays = df["owd_ms"]

        return {
            "iteration": iteration,
            "file_path": full_path,
            "count": len(delays),
            "mean_owd_ms": delays.mean(),
            "std_owd_ms": delays.std(),
            "median_owd_ms": delays.median(),
            "p99_owd_ms": np.percentile(delays, 99),
        }
    except Exception as e:
        print(f"Error processing {full_path}: {e}")
        return None

# === FIND ALL CSV FILES WITH ITERATION INDEX ===
csv_files = []
iteration = 0
for dirpath, _, filenames in os.walk(root_dir):
    for file in filenames:
        if file.endswith(".csv"):
            full_path = os.path.join(dirpath, file)
            csv_files.append((iteration, full_path))
            iteration += 1

# === MULTITHREADED PROCESSING ===
summary_stats = []

with ThreadPoolExecutor() as executor:
    futures = {executor.submit(process_csv, file_info): file_info for file_info in csv_files}
    for future in tqdm(as_completed(futures), total=len(futures), desc="Processing CSVs"):
        result = future.result()
        if result:
            summary_stats.append(result)

# === SAVE TO CSV ===
summary_df = pd.DataFrame(summary_stats)
summary_df.sort_values(by="iteration", inplace=True)
summary_df.to_csv(output_file, index=False)
print(f"Saved summary to {output_file}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# === CONFIGURE FOR ACADEMIC QUALITY PLOTS ===
rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 13,
    "axes.titlesize": 14,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "pdf.fonttype": 42,
})

# === LOAD DATA ===
df = pd.read_csv("delay_statistics_summary-Tor.csv")
valid_p99 = df["p99_owd_ms"]
valid_p99 = valid_p99[valid_p99 > 0]  # Ignore non-positive values

# === REMOVE OUTLIERS USING IQR METHOD ===
Q1 = valid_p99.quantile(0.25)
Q3 = valid_p99.quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

filtered = valid_p99[valid_p99 <= upper_bound]
ignored_count = len(valid_p99) - len(filtered)
ignored_percentage = (ignored_count / len(valid_p99)) * 100

# === PERCENTAGE ABOVE THRESHOLDS ===
above_360 = (filtered > 360).sum()
above_400 = (filtered > 400).sum()
percent_above_360 = (above_360 / len(filtered)) * 100
percent_above_400 = (above_400 / len(filtered)) * 100

print(f"Ignored {ignored_count} outliers out of {len(valid_p99)} total ({ignored_percentage:.2f}%)")
print(f"{percent_above_360:.2f}% of values are above 360 ms")
print(f"{percent_above_400:.2f}% of values are above 400 ms")

# === CDF PLOT ===
sorted_vals = np.sort(filtered)
cdf = np.arange(1, len(sorted_vals)+1) / len(sorted_vals)

plt.figure(figsize=(8, 5))
plt.plot(sorted_vals, cdf, marker='.', linestyle='-', alpha=0.8)
plt.axvline(x=400, color='red', linestyle='--', linewidth=1.5, label="400 ms Threshold")
plt.xlabel("99th Percentile Delay (ms)")
plt.ylabel("CDF")
plt.title("CDF of 99th Percentile of RTP Delays (Outliers Removed)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("cdf_99th_percentile_delay_threshold.pdf", dpi=300, bbox_inches="tight", transparent=True)
plt.show()
```
