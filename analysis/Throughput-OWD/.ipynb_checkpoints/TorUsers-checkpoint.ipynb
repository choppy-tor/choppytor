```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

# === CONFIGURE FOR ACADEMIC QUALITY PLOTS ===
rcParams.update({
    "font.family": "serif",
    "font.size": 12,
    "axes.labelsize": 13,
    "axes.titlesize": 14,
    "legend.fontsize": 11,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "pdf.fonttype": 42,  # Editable fonts
})

# === READ CSV FILE ===
file_path = "userstats-relay-country-2018-01-01-2025-05-16-all-off.csv"  # Update path if needed
df = pd.read_csv(file_path, sep=",")  # Use '\t' if tab-separated

# === CLEAN & CONVERT DATE ===
df['date'] = pd.to_datetime(df['date'])
df = df.dropna(subset=['users'])  # Remove rows where 'users' is missing

# === PLOT ===
plt.figure(figsize=(6, 3.5))  # Suitable for 1-column academic plot
plt.plot(df['date'], df['users'], label='Tor Users', linewidth=1.5)

plt.xlabel("Date")
plt.ylabel("Number of Users")
# plt.title("Daily Tor User Trends")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.legend()

# === SAVE PLOT ===
plt.savefig("tor_user_trends.pdf", bbox_inches='tight')
plt.show()
```
