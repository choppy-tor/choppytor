```python
import os
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# === File Paths ===
base_dir = "/home/jithin/data/IMC/NewData/Tor"
tor_throughput_file = "onionperf-throughput-2025-04-01-2025-08-25-public.csv"

# === Extract Measured Throughput from log.json ===
throughput_values = []

for root, _, files in os.walk(base_dir):
    for file in files:
        if file == "log.json":
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r') as f:
                    data = json.load(f)
                    for entry in data:
                        try:
                            if "iperf" in entry and "receiver" in entry["iperf"]:
                                bitrate = entry["iperf"]["receiver"].get("bitrate", "")
                                unit = entry["iperf"]["receiver"].get("unit", "")
                                if bitrate and unit:
                                    bitrate = float(bitrate)
                                    # Convert all to Kbits/sec
                                    if "Mbit" in unit:
                                        bitrate *= 1000
                                    elif "Gbit" in unit:
                                        bitrate *= 1000000
                                    throughput_values.append(bitrate)
                        except:
                            continue
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

measured_throughput = np.array(throughput_values)
measured_throughput = measured_throughput[measured_throughput > 0]



# === Extract Tor Official Throughput ===
try:
    tor_df = pd.read_csv(tor_throughput_file)
    if {'date', 'md'}.issubset(tor_df.columns):
        tor_throughput = tor_df['md'].dropna().astype(float).values
        tor_throughput = tor_throughput[tor_throughput > 0]
    else:
        raise ValueError("Missing required column 'md'")
except Exception as e:
    print(f"Error reading Tor throughput file: {e}")
    tor_throughput = []

print("Length of measured throughput: ", len(measured_throughput))
print("Length of tor throughput: ", len(tor_throughput))
# Create a CDF for each group
def compute_cdf(data):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    return sorted_data, cdf

# Compute CDFs
measured_sorted, measured_cdf = compute_cdf(measured_throughput)
print(len(measured_sorted))
tor_sorted, tor_cdf = compute_cdf(tor_throughput) if len(tor_throughput) > 0 else ([], [])
print(len(tor_sorted))

# Recalculate percentages
measured_above_500 = np.sum(measured_throughput > 500)
measured_total = len(measured_throughput)
measured_percent_above_500 = (measured_above_500 / measured_total) * 100 if measured_total > 0 else 0

tor_above_500 = np.sum(tor_throughput > 500)
tor_total = len(tor_throughput)
tor_percent_above_500 = (tor_above_500 / tor_total) * 100 if tor_total > 0 else 0

print(measured_percent_above_500)
print(tor_percent_above_500)


# === Plot CDF ===
fig, ax = plt.subplots(figsize=(3.5, 2.5))  # 1-column width

ax.plot(measured_sorted, measured_cdf, label="Measured Throughput", linewidth=1.5)
if len(tor_sorted) > 0:
    ax.plot(tor_sorted, tor_cdf, label="Tor Official Throughput", linewidth=1.5)

# === Mark 400 ms Threshold ===
ax.axvline(x=500, linestyle='--', color='blue', linewidth=1.0)

ax.set_xlabel("Throughput (Kbps)")
ax.set_ylabel("CDF")
ax.grid(True, linestyle='--', linewidth=0.6, alpha=0.6)
ax.legend(loc="lower right", fontsize=8)

plt.tight_layout()
plt.savefig("throughput_cdf_plot.pdf", dpi=300, bbox_inches="tight", transparent=True)
plt.show()

# # === Prepare Data for Plotting ===
# df_plot = pd.DataFrame({
#     "Throughput (Kbps)": np.concatenate([measured_throughput, tor_throughput]),
#     "Source": (["Measured Throughput"] * len(measured_throughput)) + (["Tor Official Throughput"] * len(tor_throughput))
# })

# # === Styling ===
# sns.set_theme(style="whitegrid", font_scale=0.9, rc={
#     "axes.labelsize": 9,
#     "xtick.labelsize": 8,
#     "ytick.labelsize": 8,
#     "lines.linewidth": 1.2
# })

# # === Plot ===
# fig, ax = plt.subplots(figsize=(3.5, 2.5))  # 1-column width

# sns.boxplot(
#     data=df_plot,
#     x="Source",
#     y="Throughput (Kbps)",
#     ax=ax,
#     palette=["lightblue", "lightgreen"],
#     width=0.5,
#     showfliers=False  # <<< REMOVE OUTLIERS
# )

# # === Clean up ===
# ax.set_xlabel("")
# ax.set_ylabel("Throughput (Kbps)")
# ax.set_ylim(0, df_plot["Throughput (Kbps)"].quantile(0.95))  # Trim at 95th percentile for visibility
# ax.grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.6)

# plt.tight_layout()
# plt.savefig("throughput_boxplot_seaborn_no_outliers.pdf", dpi=300, bbox_inches="tight", transparent=True)
# plt.show()
```
